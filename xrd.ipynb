```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import *
from pymatgen.core.operations import *
from pymatgen import Structure
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
with MPRester("TD3CpcK0oJ60Lhc1ZD") as m:
    struct = m.get_structure_by_material_id("mp-754825")
```

```python
xrd_cal = XRDCalculator('CuKa')
diseno = xrd_cal.get_pattern(struct)
```

```python
print("Premier pic du diffractogramme :")
print(" ")
print("angle 2*theta [°] ", diseno.x[0])
print("intensité : [%]", diseno.y[0])
print("indices hkl associés: ", diseno.hkls[0])
```

```python
print("Deuxième pic du diffractogramme :")
print(" ")
print("angle 2*theta [°] ", diseno.x[1])
print("intensité : [%]", diseno.y[1])
print("indices hkl associés: ", diseno.hkls[1])
```

```python
print("Troisième pic du diffractogramme :")
print(" ")
print("angle 2*theta [°] ", diseno.x[2])
print("intensité : [%]", diseno.y[2])
print("indices hkl associés: ", diseno.hkls[2])
```

```python
print("Trois premiers pics du diffractogramme")
xrd_cal.show_plot(struct,two_theta_range=(10,32.5),annotate_peaks=True,ax=None,with_labels=True)
```
